# What is hypothesis testing in statistics
Hypothesis testing is a statistical method used to determine if there enough evidence in a sample to support a claim about a population.

# What is the null hypothesis, and how does it differ from the alternative hypothesis
The null hypothesis   proposes that there is no significant relationship or difference between variables while the alternative hypothesis  suggests that there is a significant relationship or difference.

# What is the significance level in hypothesis testing, and why is it important

hypothesis testing, the significance level, denoted by Î±  is the probability of rejecting the null hypothesis when it is actually true.
Importance of Significance Level Controls for Type I Errors
Determines Statistical Significance.

# What does a P-value represent in hypothesis testing
 hypothesis testing a p-value represents the probability of obtaining results as extreme as or more extreme than the observed results assuming the null hypothesis is true.

# How do you interpret the P-value in hypothesis testing
hypothesis testing, the p-value represents the probability of observing results as extreme as or more extreme than the ones observed in the study assuming the null hypothesis is true.A p-value measures the probability of obtaining the observed results assuming that the null hypothesis is true. The lower the p-value the greater the statistical significance of the observed difference.


# What are Type 1 and Type 2 errors in hypothesis testing.
a Type I error false positive occurs when you reject a true null hypothesis, and a Type II error false negative occurs when you fail to reject a false null hypothesis.
Type I Error False Positive
This error happens when you reject the null hypothesis  when it is actually true. In other words  you conclude there  an effect or relationship when there isnt one.
Type II Error False Negative
This error occurs when you fail to reject the null hypothesis when it is actually false. You conclude that there no effect or relationship when one actually exists.


# What is the difference between a one-tailed and a two-tailed test in hypothesis testing
a one-tailed test checks for a difference in one specific direction.while a two-tailed test checks for a difference in either direction

# What is the Z-test, and when is it used in hypothesis testing
z-test is a statistical method used to test hypotheses about population parameters particularly the population mean when the population standard deviation is known or when the sample size is large enough   to apply the Central Limit Theorem.

# How do you calculate the Z-score, and what does it represent in hypothesis testing
 the Z-score is represents how many standard deviations a particular data point is from the mean of a dataset. It is calculate by subtracting the mean from the data point and then dividing by the standard deviation.

# What is the T-distribution, and when should it be used instead of the normal distribution
The t-distribution is a probability distribution similar to the normal distribution but with heavier tails meaning it has a greater chance of extreme values. It's used when dealing with small sample sizes or when the population standard deviation is unknown.

# What is the difference between a Z-test and a T-test
The main difference between a Z-test and a T-test lies in the assumption about the population standard deviation and the sample size used in the test.

# What is the T-test, and how is it used in hypothesis testing.
A t-test is a statistical hypothesis test used to determine if there's a significant difference between the means of two groups.

# What is the relationship between Z-test and T-test in hypothesis testing
The main difference is that a t-test is used for small sample sizes n <30 or when the population variance is unknown and uses the t-distribution. A Z-test is used for large sample sizes n>30 with known population variance and relies on the normal distribution.

# What is a confidence interval, and how is it used to interpret statistical results
A confidence interval is a range of values within which the true population parameter is likely to fall based on a sample.

# What is the margin of error, and how does it affect the confidence interval
The margin of error is a statistical measure that reflects the precision of an estimate essentially how much the sample results might differ from the real population results.

# How is Bayes Theorem used in statistics, and what is its significance
Bayes Theorem is a fundamental statistical concept used to update the probability of a hypothesis based on new evidence.

# What is the Chi-square distribution, and when is it used
The chi-square distribution is a continuous probability distribution commonly used in hypothesis testing particularly the chi-square goodness-of-fit test and the chi-square test of independence
when is it used
Goodness-of-fit test >This test determines how well a sample observed frequencies match a theoretical distribution.
test of independence> This test determines if there a relationship between two categorical variables.

# What is the Chi-square goodness of fit test, and how is it applied
The chi-square goodness of fit test is a statistical test used to determine how well a sample data matches a specific distribution or population
how is it  applied
The null hypothesis assumes that the observed data comes from the specified distribution. The alternative hypothesis suggests that the observed data does not come from the specified distribution.
Based on the hypothesized distribution, determine the expected frequency for each category or class.

# > What is the F-distribution, and when is it used in hypothesis testing
The F-distribution is a continuous probability distribution used in hypothesis testing particularly when comparing variances between two or more samples.

# What is an ANOVA test, and what are its assumptions
ANOVA Analysis of Variance is a statistical test used to compare the means of two or more groups Its primary goal is to determine if there a statistically significant difference between the group means or if the observed differences are due to random chance. ANOVA relies on several assumptions that must be met for the results to be valid.
Each observation within and between groups should be independent of each other.
The variance within each group should be approximately equal.


# What are the different types of ANOVA tests
There are two main types of ANOVA tests one-way ANOVA and two-way ANOVA. One-way ANOVA compares the means of three or more groups based on a single independent variable while two-way ANOVA analyzes the effects of two or more independent variables on a dependent variable.

# What is the F-test, and how does it relate to hypothesis testing?
The F-test is a statistical test used to compare variances between two or more groups.
It often part of the analysis of variance (ANOVA) and is used to determine if there's a significant difference between the means of those groups.

```python
#Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
def bernoulli_pdf(k, p):
  """
  Calculates the Probability Distribution Function (PDF) of a Bernoulli distribution.

  Args:
    k: The value at which to evaluate the PDF (0 or 1).
    p: The probability of success (a value between 0 and 1).

  Returns:
    The probability of observing the value k for a Bernoulli distribution with parameter p.
    Returns 0 if k is not 0 or 1.
  """
  if k == 1:
    return p
  elif k == 0:
    return 1 - p
  else:
    return 0

p_success = 0.7
print(f"PDF at k=1 (success) for p={p_success}: {bernoulli_pdf(1, p_success)}")
print(f"PDF at k=0 (failure) for p={p_success}: {bernoulli_pdf(0, p_success)}")
print(f"PDF at k=2 for p={p_success}: {bernoulli_pdf(2, p_success)}")
```

```python
# prompt:  Write a Python function to calculate Z-scores from a dataset and plot them

def calculate_and_plot_z_scores(data):
  """
  Calculates Z-scores for a dataset and plots a histogram of the Z-scores.

  Args:
    data: A list or NumPy array of numerical data.
  """
  mean = np.mean(data)
  std_dev = np.std(data)

  if std_dev == 0:
    print("Standard deviation is zero. Z-scores cannot be calculated.")
    return

  z_scores = (data - mean) / std_dev

  plt.figure(figsize=(8, 6))
  plt.hist(z_scores, bins=30, density=True, alpha=0.7)
  plt.xlabel('Z-score')
  plt.ylabel('Density')
  plt.title('Histogram of Z-scores')
  plt.grid(axis='y', alpha=0.5)
  plt.show()

calculate_and_plot_z_scores(normal_data)
```

```python
#Implement hypothesis testing using Z-statistics for a sample dataset

from statsmodels.stats.weightstats import ztest
sample_data = np.array([102, 105, 98, 103, 101, 100, 104, 106, 99, 102])
hypothesized_mean = 100
population_std_dev = np.std(sample_data, ddof=0)
z_statistic, p_value = ztest(sample_data, value=hypothesized_mean, alternative='two-sided')
print(f"Sample Data: {sample_data}")
print(f"Hypothesized Mean (Î¼â): {hypothesized_mean}")
print(f"Assumed Population Standard Deviation (Ï): {population_std_dev:.4f}")
print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
alpha = 0.05
print(f"\nSignificance Level (Î±): {alpha}")
if p_value < alpha:
  print("Result: Reject the null hypothesis.")
  print("Conclusion: There is statistically significant evidence to suggest that the true population mean is different from the hypothesized mean.")
else:
  print("Result: Fail to reject the null hypothesis.")
  print("Conclusion: There is not enough statistically significant evidence to suggest that the true population mean is different from the hypothesized mean.")
```

```python
#Create a confidence interval for a dataset using Python and interpret the result

from scipy.stats import norm

dataset = np.array([12, 15, 11, 18, 14, 17, 13, 16, 10, 19])


sample_mean = np.mean(dataset)
sample_std = np.std(dataset, ddof=1)  # Use ddof=1 for sample standard deviation
n = len(dataset)
standard_error = sample_std / np.sqrt(n)
confidence_level = 0.95
alpha = 1 - confidence_level
z_critical = norm.ppf(1 - alpha/2)
margin_of_error = z_critical * standard_error
confidence_interval_lower = sample_mean - margin_of_error
confidence_interval_upper = sample_mean + margin_of_error
print(f"Dataset: {dataset}")
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Standard Deviation: {sample_std:.4f}")
print(f"Sample Size (n): {n}")
print(f"Standard Error: {standard_error:.4f}")
print(f"Confidence Level: {confidence_level}")
print(f"Z-critical value: {z_critical:.4f}")
print(f"Margin of Error: {margin_of_error:.4f}")
print(f"Confidence Interval: ({confidence_interval_lower:.4f}, {confidence_interval_upper:.4f})")
print("\nInterpretation:")
print(f"We are {confidence_level*100}% confident that the true population mean")
print(f"lies within the range of ({confidence_interval_lower:.4f}, {confidence_interval_upper:.4f}).")
print("This means that if we were to take many random samples from the same population")
print("and calculate a confidence interval for each, approximately 95% of these intervals")
print("would contain the true population mean.")
```

```python
#Implement Z-score normalization on a dataset and visualize its transformation

def z_score_normalization(data):
  """
  Performs Z-score normalization on a dataset.

  Args:
    data: A NumPy array of numerical data.

  Returns:
    A NumPy array containing the Z-scores of the input data.
    Returns None if the standard deviation is zero.
  """
  mean = np.mean(data)
  std_dev = np.std(data)
  if std_dev == 0:
    print("Standard deviation is zero. Z-score normalization cannot be performed.")
    return None

  z_normalized_data = (data - mean) / std_dev
  return z_normalized_data

sample_data_for_normalization = np.random.normal(loc=50, scale=15, size=1000)
normalized_data = z_score_normalization(sample_data_for_normalization)

if normalized_data is not None:
  plt.figure(figsize=(12, 5))
  plt.subplot(1, 2, 1)
  plt.hist(sample_data_for_normalization, bins=30, density=True, alpha=0.7, color='skyblue')
  plt.xlabel('Original Value')
  plt.ylabel('Density')
  plt.title('Original Data Distribution')
  plt.grid(axis='y', alpha=0.5)
  plt.subplot(1, 2, 2)
  plt.hist(normalized_data, bins=30, density=True, alpha=0.7, color='lightcoral')
  plt.xlabel('Z-score')
  plt.ylabel('Density')
  plt.title('Z-score Normalized Data Distribution')
  plt.grid(axis='y', alpha=0.5)
  plt.tight_layout()
  plt.show()
  print("Original Data - Mean:", np.mean(sample_data_for_normalization))
  print("Original Data - Standard Deviation:", np.std(sample_data_for_normalization))
  print("Normalized Data - Mean:", np.mean(normalized_data))
  print("Normalized Data - Standard Deviation:", np.std(normalized_data))
```

```python
#Perform a two-tailed Z-test using Python and visualize the decision region on a plot

def plot_two_tailed_z_test(sample_mean, hypothesized_mean, std_dev, n, alpha):
  """
  Performs a two-tailed Z-test and visualizes the decision regions.

  Args:
    sample_mean: The mean of the sample.
    hypothesized_mean: The mean specified in the null hypothesis (Î¼â).
    std_dev: The population standard deviation (Ï).
    n: The sample size.
    alpha: The significance level.
  """
  standard_error = std_dev / np.sqrt(n)
  z_statistic = (sample_mean - hypothesized_mean) / standard_error
  p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

  print(f"\n--- Two-tailed Z-test Results ---")
  print(f"Sample Mean: {sample_mean:.4f}")
  print(f"Hypothesized Mean (Î¼â): {hypothesized_mean}")
  print(f"Population Standard Deviation (Ï): {std_dev:.4f}")
  print(f"Sample Size (n): {n}")
  print(f"Standard Error of the Mean: {standard_error:.4f}")
  print(f"Z-statistic: {z_statistic:.4f}")
  print(f"P-value: {p_value:.4f}")
  print(f"Significance Level (Î±): {alpha}")


  z_critical_lower = norm.ppf(alpha / 2)
  z_critical_upper = norm.ppf(1 - alpha / 2)

  print(f"Critical Z-values: ({z_critical_lower:.4f}, {z_critical_upper:.4f})")

  print("\nInterpretation:")
  if abs(z_statistic) > z_critical_upper: # or p_value < alpha
    print("Result: Reject the null hypothesis (Hâ).")
    print("Conclusion: There is statistically significant evidence at the {alpha*100}% significance level to suggest that the true population mean is different from the hypothesized mean.")
  else:
    print("Result: Fail to reject the null hypothesis (Hâ).")
    print("Conclusion: There is not enough statistically significant evidence at the {alpha*100}% significance level to suggest that the true population mean is different from the hypothesized mean.")

  x = np.linspace(-4, 4, 1000)
  pdf_values = norm.pdf(x, loc=0, scale=1)

  plt.figure(figsize=(10, 6))
  plt.plot(x, pdf_values, label='Standard Normal Distribution')


  x_reject_lower = x[x <= z_critical_lower]
  plt.fill_between(x_reject_lower, 0, norm.pdf(x_reject_lower, loc=0, scale=1), color='red', alpha=0.3, label='Rejection Region (Î±/2)')

  x_reject_upper = x[x >= z_critical_upper]
  plt.fill_between(x_reject_upper, 0, norm.pdf(x_reject_upper, loc=0, scale=1), color='red', alpha=0.3)

  plt.axvline(z_statistic, color='blue', linestyle='dashed', linewidth=2, label=f'Observed Z-statistic = {z_statistic:.2f}')

  plt.xlabel('Z-score')
  plt.ylabel('Density')
  plt.title('Two-tailed Z-test: Decision Regions on Standard Normal Distribution')
  plt.grid(True)
  plt.legend()
  plt.show()
```

```python
#Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

def visualize_type1_type2_errors(mu0, mu_a, sigma, n, alpha):
  """
  Calculates and visualizes Type 1 and Type 2 errors for a one-tailed Z-test.

  Args:
    mu0: The mean under the null hypothesis (H0).
    mu_a: The true mean under the alternative hypothesis (H1).
    sigma: The population standard deviation.
    n: The sample size.
    alpha: The significance level (Type 1 error rate).
  """
  std_error = sigma / np.sqrt(n)


  z_critical_alpha = norm.ppf(1 - alpha)
  critical_sample_mean_alpha = mu0 + z_critical_alpha * std_error


  z_beta = (critical_sample_mean_alpha - mu_a) / std_error
  beta = norm.cdf(z_beta)

  print(f"\n--- Type 1 and Type 2 Error Visualization ---")
  print(f"Null Hypothesis (H0): Î¼ = {mu0}")
  print(f"Alternative Hypothesis (H1): Î¼ > {mu0}")
  print(f"True Population Mean (Î¼_a): {mu_a}")
  print(f"Population Standard Deviation (Ï): {sigma}")
  print(f"Sample Size (n): {n}")
  print(f"Significance Level (Î±): {alpha:.4f}")
  print(f"Critical Z-value (Î±): {z_critical_alpha:.4f}")
  print(f"Critical Sample Mean (for rejection): {critical_sample_mean_alpha:.4f}")
  print(f"Beta (Type 2 Error Rate): {beta:.4f}")
  print(f"Power (1 - Beta): {1 - beta:.4f}")

  x = np.linspace(min(mu0, mu_a) - 4 * std_error, max(mu0, mu_a) + 4 * std_error, 1000)

  pdf_h0 = norm.pdf(x, loc=mu0, scale=std_error)
  plt.figure(figsize=(10, 6))
  plt.plot(x, pdf_h0, label=f'Distribution under H0 (Î¼={mu0})')

  pdf_h1 = norm.pdf(x, loc=mu_a, scale=std_error)
  plt.plot(x, pdf_h1, label=f'Distribution under H1 (Î¼={mu_a})', linestyle='dashed')

  x_type1_error = x[x >= critical_sample_mean_alpha]
  plt.fill_between(x_type1_error, 0, norm.pdf(x_type1_error, loc=mu0, scale=std_error), color='red', alpha=0.3, label=f'Type I Error (Î± = {alpha:.2f})')

  x_type2_error = x[x < critical_sample_mean_alpha]
  plt.fill_between(x_type2_error, 0, norm.pdf(x_type2_error, loc=mu_a, scale=std_error), color='orange', alpha=0.3, label=f'Type II Error (Î² = {beta:.2f})')

  plt.axvline(critical_sample_mean_alpha, color='black', linestyle='dotted', linewidth=2, label=f'Critical Sample Mean = {critical_sample_mean_alpha:.2f}')

  plt.xlabel('Sample Mean')
  plt.ylabel('Density')
  plt.title('Visualization of Type I and Type II Errors (One-tailed test)')
  plt.grid(True)
  plt.legend()
  plt.show()
```

```python
#implement a Bayesian inference method using Bayes'Theorem in Python and explain the process


def bayesian_inference(prior_probability, likelihood, evidence):
    """
    Calculates the posterior probability using Bayes' Theorem.

    Bayes' Theorem: P(Hypothesis | Evidence) = [P(Evidence | Hypothesis) * P(Hypothesis)] / P(Evidence)

    Args:
        prior_probability (float): The prior probability of the hypothesis P(Hypothesis).
                                   This is your initial belief before observing the evidence.
        likelihood (float): The likelihood of observing the evidence given the hypothesis is true P(Evidence | Hypothesis).
                            This comes from your model of how the evidence is generated under the hypothesis.
        evidence (float): The probability of the evidence itself P(Evidence).
                          This can be calculated as P(Evidence | Hypothesis) * P(Hypothesis) + P(Evidence | not Hypothesis) * P(not Hypothesis).
                          In many simple examples, P(Evidence) might be implicitly handled or provided.

    Returns:
        float: The posterior probability of the hypothesis given the evidence P(Hypothesis | Evidence).
               Returns None if the evidence probability is zero.
    """
    if evidence == 0:
        print("Evidence probability is zero. Cannot perform Bayesian inference.")
        return None

    posterior_probability = (likelihood * prior_probability) / evidence
    return posterior_probability

print("--- Bayesian Inference using Bayes' Theorem ---")
print("Bayes' Theorem allows us to update our belief about a hypothesis based on new evidence.")
print("The core components are:")
print("1. Prior Probability (P(Hypothesis)): Our initial belief in the hypothesis before seeing any evidence.")
print("2. Likelihood (P(Evidence | Hypothesis)): The probability of observing the evidence if the hypothesis is true.")
print("3. Evidence (P(Evidence)): The overall probability of observing the evidence, regardless of the hypothesis.")
print("   This acts as a normalizing factor.")
print("4. Posterior Probability (P(Hypothesis | Evidence)): Our updated belief in the hypothesis after seeing the evidence.")
print("\nThe formula is: P(Hypothesis | Evidence) = [P(Evidence | Hypothesis) * P(Hypothesis)] / P(Evidence)")
print("In essence, the posterior probability is proportional to the likelihood times the prior probability.")
print("P(Hypothesis | Evidence) â P(Evidence | Hypothesis) * P(Hypothesis)")
print("The evidence P(Evidence) ensures the probabilities sum to 1.")
disease_prevalence = 0.001
test_sensitivity = 0.99
prob_no_disease = 1 - disease_prevalence
false_positive_rate = 0.05
evidence_prob = (test_sensitivity * disease_prevalence) + (false_positive_rate * prob_no_disease)
print("\n--- Example Calculation ---")
print(f"Prior Probability (Disease Prevalence P(H)): {disease_prevalence}")
print(f"Likelihood (Test Sensitivity P(E | H)): {test_sensitivity}")
print(f"False Positive Rate P(E | not H): {false_positive_rate}")
print(f"Probability of No Disease P(not H): {prob_no_disease}")
print(f"Probability of Evidence (Positive Test) P(E): {evidence_prob:.6f}")
posterior = bayesian_inference(prior_probability=disease_prevalence,
                               likelihood=test_sensitivity,
                               evidence=evidence_prob)
if posterior is not None:
    print(f"\nPosterior Probability P(H | E) (Probability of having the disease given a positive test): {posterior:.4f}")

print("\n--- Interpretation ---")
print(f"Starting with a low prior belief ({disease_prevalence*100:.2f}%), a positive test")
print(f"(even with a good sensitivity and specificity) updates our belief to a posterior probability")
print(f"of {posterior*100:.2f}%. This demonstrates how evidence shifts our belief from the prior to the posterior.")
print("The posterior probability is the updated belief in the hypothesis after incorporating the evidence.")
```

```python
#Perform a Chi-square test for independence between two categorical variables in Python

from scipy.stats import chi2_contingency


observed_data = np.array([[50, 30],
                          [40, 70]])

print("Observed Contingency Table:")
print(observed_data)

chi2_statistic, p_value, dof, expected_frequencies = chi2_contingency(observed_data)

print("\nChi-square Test for Independence Results:")
print("-" * 40)
print(f"Chi-square Statistic: {chi2_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")
print("\nExpected Frequencies (assuming independence):")
print(expected_frequencies)


alpha = 0.05

print(f"\nSignificance Level (Î±): {alpha}")

print("\nInterpretation:")
print("Null Hypothesis (Hâ): The two categorical variables are independent.")
print("Alternative Hypothesis (Hâ): The two categorical variables are not independent (there is a relationship).")

if p_value < alpha:
  print("\nResult: Reject the null hypothesis (Hâ).")
  print(f"Conclusion: There is statistically significant evidence at the {alpha*100}% significance level to suggest that there is a relationship between the two categorical variables.")
else:
  print("\nResult: Fail to reject the null hypothesis (Hâ).")
  print(f"Conclusion: There is not enough statistically significant evidence at the {alpha*100}% significance level to suggest that there is a relationship between the two categorical variables.")
```

```python
#Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
#data.

def calculate_expected_frequencies(observed_data):
  """
  Calculates the expected frequencies for a Chi-square test based on observed data.

  Args:
    observed_data: A NumPy array representing the observed contingency table.

  Returns:
    A NumPy array containing the expected frequencies.
  """
  row_sums = np.sum(observed_data, axis=1)
  col_sums = np.sum(observed_data, axis=0)
  total_sum = np.sum(observed_data)

  expected_frequencies = np.outer(row_sums, col_sums) / total_sum

  return expected_frequencies

observed_data_example = np.array([[50, 30],
                                  [40, 70]])

expected_freq = calculate_expected_frequencies(observed_data_example)

print("Observed Contingency Table:")
print(observed_data_example)

print("\nCalculated Expected Frequencies:")
print(expected_freq)

chi2_statistic, p_value, dof, expected_frequencies_scipy = chi2_contingency(observed_data_example)
print("\nExpected Frequencies from scipy.stats.chi2_contingency:")
expected_frequencies_scipy
```
